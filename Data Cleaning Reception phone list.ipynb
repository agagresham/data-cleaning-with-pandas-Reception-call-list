```python
# looking and drop duplicates 
df =df.drop_duplicates()
```

```python
# remove all columns what we do not need
df =df.drop(columns = 'Not_Useful_Column')
```

```python
# clean last name column, right, left strip or total strip as below 
#df['Last_Name']=df['Last_Name'].str.lstrip('...')
#df['Last_Name']=df['Last_Name'].str.lstrip('/')
#df['Last_Name']=df['Last_Name'].str.rstrip('_')
df['Last_Name']=df['Last_Name'].str.strip('123._/')
```

```python
# cleaning phone numbers in diffrent formats, strip first to numbers only and coming back to reformating as we want to
# this will replacing all string from small carracters a-z,then from big carracters A-Z,and numbers 0-9
#df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex = True)
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex = True)
```

```python
# before formating we need to change type into the string as in data are NaN or Na values
# we can apply this using lambda
df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number']= df['Phone_Number'].apply(lambda x: x[0:3] + '-'+x[3:6] + '-'+x[6:10])
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('Na--','')
```

```python
# looking on adress column and split into columns
df[['Street_adress','State','Zip_code']] = df['Address'].str.split(',',expand = True)
```

```python
df
```

```python
#looking at Paying customer, street_adress of na and do not contact column
df['Address']= df['Address'].str.replace('N/a','')
df['Street_adress']= df['Street_adress'].str.replace('N/a','')
df['Paying Customer']= df['Paying Customer'].str.replace('N/a','')
df['Paying Customer']= df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']= df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('N/a','')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes','Y')
```

```python
df
```

```python
# removing all empty na data
df = df.fillna('')
```

```python
df
```

```python
# removing from list all customers what have y
for x in df.index:
   if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace = True)
    
df
```

```python
# if phone is blank remove 
for x in df.index:
   if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace = True)

df
```

```python
# reset index from 0 to 7 this showing final product and we can handle list to reception 
df = df.reset_index(drop = True)
```
